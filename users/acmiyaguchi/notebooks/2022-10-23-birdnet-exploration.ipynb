## embedding analysis

```python
! mkdir -p ../data/processed/birdnet/embedding/afrsil1
! rm -r ../data/processed/birdnet/embedding/afrsil1

! docker run --rm \
    -v ${PWD}/../data/raw/birdclef-2022/train_audio/afrsil1:/input \
    -v ${PWD}/../data/processed/birdnet/embedding/afrsil1:/output \
    -it us-central1-docker.pkg.dev/birdclef-eda-f22/birdclef-eda-f22/birdnet:latest \
    embeddings.py \
        --i /input \
        --o /output \
        --threads 8 \
        --batchsize 16
```

```python
filename = "../data/processed/birdnet/embedding/afrsil1/XC205893.birdnet.embeddings.txt"
df_emb = pd.read_csv(filename, sep="\t", header=None)
df_emb.columns = ["start", "end", "embedding"]
# parse embedding column from a comma-delimited string into a list of floats
df_emb["embedding"] = df_emb["embedding"].apply(
    lambda x: [float(i) for i in x.split(",")]
)
df_emb.head()
```

```python
# NOTE: we know that the african silverbill is always the highest confidence
confidences = []
for time in df["Start (s)"].unique():
    subset = df[df["Start (s)"] == time]
    confidences.append(subset["Confidence"].max())
```

```python
import numpy as np
from sklearn.decomposition import PCA

X = np.stack(df_emb["embedding"].values)

pca = PCA(n_components=2)
G = pca.fit_transform(X)


plt.title("PCA of BirdNET embeddings for African Silverbill")
plt.scatter(G[:, 0], G[:, 1], c=confidences)
plt.colorbar()
plt.show()
```
